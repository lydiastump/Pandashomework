```python
school_data5 = pd.merge(school_data4, passingreadingschool_df, on=['School ID'], how='inner')
school_data5.head()

school_data5['% Passing reading'] = school_data5['Number passing reading']/school_data5['size']*100
del school_data5['Number passing reading']
```
